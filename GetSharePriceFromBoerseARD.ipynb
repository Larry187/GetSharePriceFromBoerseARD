```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://kurse.boerse.ard.de/ard/kurse_einzelkurs_uebersicht.htn?i=107741'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
soup
```

```python
index_lastprice = soup.find_all('script')[14].contents[0].find('lastprice')
index_lastprice
```

```python
lastprice = soup.find_all('script')[14].contents[0][index_lastprice+12:index_lastprice+18].split('"')[0]
```

```python
name = soup.find_all('title')[0].contents[0].split('aktuell')[0]
```

```python
print("Last share price of "+name+" is: " + lastprice + "€")
```
